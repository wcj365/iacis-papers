```python
import numpy as np
import pandas as pd
import requests 
from urllib.parse import urljoin
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import re
```

```python
url = "http://www.iacis.org/iis/author_index.php"
html_content = requests.get(url).text
soup = BeautifulSoup(html_content, "lxml")
print(soup.prettify()[:4000])
```

```python
my_div = soup.find("div",{"class":"column column-650 column-left"})
print(my_div.prettify()[:2000]) 
```

```python
my_links = my_div.find_all('a')
my_links[:10]
```

```python
links = [link.get("href") for link in my_links]
links[0:200]
```

```python
full_links = [urljoin("http://iacis.org", link) for link in links]
full_links[:200]
```

```python
bad_links = [link for link in full_links if len(link) < 26]
bad_links
```

```python
good_links = [link for link in full_links if len(link) >= 26]
len(good_links)
```

```python
final_links = [link.replace("www.","") if link.startswith("http://www.iacis.org") else link for link in good_links]
final_links
```

```python
file = open("output.csv", "w")
file.write("URL,YEAR\n")
for link in final_links:
    file.write(r'"'+ link + r'",')
    file.write(link[21:25])
    file.write("\n")
file.close()
```
